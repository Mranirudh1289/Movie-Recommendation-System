```python
import numpy as np
import pandas as pd

movies = pd.read_csv('tmdb_5000_movies 5.csv')
credits = pd.read_csv('tmdb_5000_credits 2.csv')
```

```python
# First we merge the two table by common column :-

movies = movies.merge(credits, on='title')
```

```python
# Taking only the columns thats help in building our recoomendation system :-

movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.head()
```

```python
# Dropping the empty columns -

movies.dropna(inplace=True)
```

```python
movies.shape
```

```python
movies.iloc[0].genres
```

```python
# Making changes in our 'genre' columns :-

import ast

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L

movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
# Similarly for 'keywords' column:-

movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
# Similarly for 'cast' column but only for first three indices :-

def convert_cast(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else :
            break
    return L

movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies
```

```python
# Similarly for 'crew' column but append only director name :-

def convert_dirc(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L

movies['crew'] = movies['crew'].apply(convert_dirc)
```

```python
movies.head()
```

```python
# Making 'overview' column from string to list :-

movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# Replacing " " to "" :-

movies['genres'] = movies['genres'].apply(lambda x : [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x : [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x : [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x : [i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
# Now we concatinate the 'overview', 'keywords', 'cast', 'crew' columns into a 'tags' column :-

movies['tags'] = movies['overview'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
# Dropping all the columns and taking only these three columns :-

movies = movies[['movie_id', 'title', 'tags']]
```

```python
movies.head()
```

```python
# Converting list to string :-

movies['tags'] = movies['tags'].apply(lambda x : " ".join(x))
```

```python
movies.head()
```

```python
# Convert 'tags' column into lower case  :-

movies['tags'] = movies['tags'].apply(lambda x : x.lower())
```

```python
movies.head()
```

```python
# Using PorterStemmer to convert the similiar word into a same word like loved, loving, loved to love :-

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)

movies['tags'] = movies['tags'].apply(stem)
```

```python
# Now we will use Counter Vectorization from sklearn.feature_extraction.text :-

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
# Converting it into numpy array :-

vectors = cv.fit_transform(movies['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
# Now we are taking the distance between two movies on the basis of cosine_similarity :-

from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
# Function to fetch five similar movies :-

def recommend(movie):
    movie_index = movies[movies['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(movies.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
